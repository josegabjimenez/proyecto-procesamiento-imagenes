```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_data = nib.load('../images/1/FLAIR.nii.gz')
```

```python
image = image_data.get_fdata()
```

```python
origin_x = 100
origin_y = 100
origin_z = 1
x = 1
y = 1
z = 1
valor_medio_cluster = image[x, y, z]
tol = 50
segmentation = np.zeros_like(image)
itera = 1
point = [x,y]
tail = [point]
evaluated=[]
while True:
  punto = tail.pop(0)

  print(punto)
  
  for dx in [-x, 0, x] :
    for dy in [-y, 0, y] :
      if (punto not in(evaluated)):
        if np.abs(valor_medio_cluster - image[punto[0]+dx, punto[1]+dy, 20]) < tol :
          if((punto[0]+dx < 100) and ((punto[0]+dx) > 0) or (punto[1]+dx < 100) and ((punto[1]+dx) > 0) ):
            segmentation[punto[0]+dx, punto[1]+dy, 20] = 1
            tail.append([punto[0]+dx, punto[1]+dy])
        else :
          if((punto[0]+dx < 100) and ((punto[0]+dx) > 0) or (punto[1]+dx < 100) and ((punto[1]+dx) > 0) ):
            segmentation[origin_x+dx, origin_y+dy, 20] = 0

  valor_medio_cluster = image[segmentation == 1].mean()

  evaluated.append(punto)

  x += 1
  y += 1
  z += 1
  itera += 1
  if len(tail) == 0:
    break


```
