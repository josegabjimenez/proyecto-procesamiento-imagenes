```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_data = nib.load('../images/1/FLAIR.nii.gz')
```

```python
image = image_data.get_fdata()
```

```python
origin_x = 100
origin_y = 100
origin_z = 1
x = 1
y = 1
z = 1
valor_medio_cluster = image[x, y, z]
tol = 3
segmentation = np.zeros_like(image)
itera = 1
while True:

  for dx in [-x, 0, x] :
    for dy in [-y, 0, y] :
      for dz in [-z, 0, z] :
        if np.abs(valor_medio_cluster - image[x+dx, y+dy, 40]) < tol :
          segmentation[origin_x+dx, origin_y+dy, 40] = 1
        else :
          segmentation[origin_x+dx, origin_y+dy, 40] = 0

  valor_medio_cluster = image[segmentation == 1].mean()

  x += 1
  y += 1
  z += 1
  itera += 1
  if itera == 40:
    break


```
