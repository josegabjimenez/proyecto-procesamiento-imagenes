```python
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_data = nib.load('../../images/1/T1.nii.gz').get_fdata()
```

```python
plt.hist(image_data[image_data>10].flatten(), 100)
```

```python
mean_value = image_data[image_data > 10].mean()
standard_deviation_value = image_data[image_data > 10].std()

print(mean_value)
print(standard_deviation_value)

image_data_rescaled = (image_data - mean_value) / (standard_deviation_value)
```

```python
plt.hist(image_data_rescaled.flatten(), 100)
```
