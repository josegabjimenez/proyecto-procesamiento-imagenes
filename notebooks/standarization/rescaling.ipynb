```python
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_data = nib.load('../../images/1/T1.nii.gz').get_fdata()
```

```python
plt.hist(image_data[image_data>10].flatten(), 100)
```

```python
plt.imshow(image_data[:, :, 100])
```

```python
min_value = image_data.min()
max_value = image_data.max()

image_data_rescaled = (image_data - min_value) / (max_value - min_value)
```

```python
plt.hist(image_data_rescaled[image_data_rescaled>0.01].flatten(), 100)
```

```python
plt.imshow(image_data_rescaled[:, :, 100])
```
