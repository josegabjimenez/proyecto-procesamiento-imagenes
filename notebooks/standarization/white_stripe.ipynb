```python
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_data = nib.load('../../images/1/T1.nii.gz').get_fdata()
```

```python
plt.hist(image_data[image_data>10].flatten(), 100)
```

```python
# from scipy.signal import find_peaks
# from scipy import stats as st
import statistics as stat

# Obtener los valores mínimo y máximo de intensidad en la imagen
min_value = np.min(image_data[image_data>10])
max_value = np.max(image_data[image_data>10])

print(min_value)
print(max_value)


# Definir los rangos para el histograma utilizando los valores mínimo y máximo
bins = int(max_value - min_value + 1)


hist, bin_edges = np.histogram(image_data[image_data>20].flatten(),bins, [min_value, max_value])

mode1 = stat.mode(image_data[image_data>10])
# values, min = np.unique(image_data[image_data>mode1], return_counts=True)
min1= np.argmin(hist)

mode = stat.mode(image_data[image_data>min1])
range = max_value / 2
p50 = np.percentile(image_data, 50)
p98 = np.percentile(image_data, 98)
pfinal = p98 - p50
mode = stat.mode(pfinal)
print("MODA1:", mode1)
print("MIN1:", min1)
print("MODA:", mode)

plt.hist(image_data[image_data>10].flatten(), 100)

# # Encontrar la moda más a la derecha en el histograma
# for i in range(len(hist)-1, 0, 9):
#     if hist[i] > hist[i-1] and hist[i] > hist[i+1]:
#         break
# mode = bin_edges[i]

# # Imprimir la moda más a la derecha encontrada
# print("Moda más a la derecha:", mode)


# Visualizar el histograma y el pico de la materia blanca encontrado
plt.plot()
plt.axvline(mode, color='r', linestyle='--')
plt.show()
```
