```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_data = nib.load('FLAIR.nii.gz')
```

```python
image = image_data.get_fdata()
```

```python
x = 1
y = 1
z = 1

tol = 3
valor_medio_cluster = image[x, y, z]
segmentation = np.zeros_like(image)

for dx in [-1, 0, 1] :
  for dy in [-1, 0, 1] :
    for dz in [-1, 0, 1] :
      if np.abs(valor_medio_cluster - image[x+dx, y+dy, z+dz]) < tol :
        segmentation[x+dx, y+dy, z+dz] = 1
      else :
        segmentation[x+dx, y+dy, z+dz] = 0

valor_medio_cluster = image[segmentation == 1].mean()
```
